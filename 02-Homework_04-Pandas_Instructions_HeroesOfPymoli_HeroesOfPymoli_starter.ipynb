```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = purchase_data["Price"].mean()
num_of_purchase = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

Purchase_Analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price":average_price,\
                                        "Number of Purchases":num_of_purchase,"Total Revenue":total_revenue})
Purchase_Analysis_df["Average Price"] = Purchase_Analysis_df["Average Price"].map("${:.2f}".format)
Purchase_Analysis_df["Total Revenue"] = Purchase_Analysis_df["Total Revenue"].map("${:.2f}".format)

Purchase_Analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,45]

label_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

bin_df1 = purchase_data.loc[:,["SN","Age"]]
bin_df = bin_df1.drop_duplicates("SN")
bin_df["Age_bins"] = pd.cut(bin_df["Age"],bins, labels = label_names)

bingroup = bin_df.groupby("Age_bins")

total_count = bingroup["Age_bins"].count()
p_players = round((bingroup["Age_bins"].count()/total_players)*100,2)

Age_Demographics = pd.DataFrame({"Total Count":total_count,"Percentage of Players":p_players })
Age_Demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_df = purchase_data.loc[:,["Purchase ID","SN","Age","Price"]]
purchase_df["Age Range"] = pd.cut(purch_df["Age"],bins, labels = label_names)
purchasegr = purchase_df.groupby("Age Range")

nodup_df = purchase_data.loc[:,["SN","Age"]]
nodup_df = nodup_df.drop_duplicates("SN")
nodup_df["Age Range"] = pd.cut(nodup_df["Age"],bins, labels = label_names)
nodup_dfgr = nodup_df.groupby("Age Range")
ndup_count = nodup_dfgr["Age Range"].count()

purchase_count = purchasegr["Purchase ID"].count()
avg_purch_price = purchasegr["Price"].mean().map("${:.2f}".format)
tot_purc_val = purchasegr["Price"].sum().map("${:.2f}".format) 
avg_tot_pur_person = (purchasegr["Price"].sum()/ndup_count).map("${:.2f}".format)

purchgrdf =pd.DataFrame({"Purchase Count":purchase_count,"Average Purchase Price":avg_purch_price,\
                        "Total Purchase Value":tot_purc_val,"Avg Total Purchase per Person":avg_tot_pur_person})
purchgrdf
```
